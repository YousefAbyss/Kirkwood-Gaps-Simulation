```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("rebound/examples/test/orbits_2.txt")
```

```python
data_t = np.array(np.split(data, 7))
a = data_t[:,:,1]
e = data_t[:,:,2]
```

```python
fig, ax = plt.subplots(a.shape[0], 1, figsize=(10, 40))
for i in range(a.shape[0]):
    ax[i].scatter(a[i], e[i], s=5)
    ax[i].set_xlim(2, 4)
    ax[i].set_ylim(0, 1)
    ax[i].set_xlabel('Semi-major axis (au)')
    ax[i].set_ylabel('Eccentricity')
    ax[i].set_title('t = ' + str(int(data_t[:,0,0][i] / 365 / 100000)) + 'e5 Years')
    ax[i].vlines(x=[2.065, 2.502, 2.825, 2.958, 3.279], ymin=0, ymax=0.8, colors='red', ls='--')
plt.show()
```

```python
bins = np.linspace(2, 4, 100)
fig, ax = plt.subplots(a.shape[0], 1, figsize=(10, 80))
for i in range(a.shape[0]):
    ax[i].hist(a[i], bins, histtype="step", density=True)
    ax[i].set_xlabel('Semi-major axis (au)')
    ax[i].set_ylabel('# of asteroids')
    ax[i].set_title('t = ' + str(int(data_t[:,0,0][i] / 365 / 100000)) + 'e5 Years')
    ax[i].vlines(x=[2.065, 2.502, 2.825, 2.958, 3.279], ymin=0, ymax=1, colors='red', ls='--')
plt.show()
```
