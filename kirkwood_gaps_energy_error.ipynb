```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
data = np.loadtxt("orbits_energy.txt")
time = data[:,0][1:] / 365 / 1e5
energy = data[:,1][1:]
```

```python
plt.plot(time, energy, 'o')
plt.xlabel('Time (years * 1e5)')
plt.ylabel('Relative Energy Error')
plt.show()
```
